# Climate Change Figures
In this notebook i will produce figures pertaining to five different aspects of climate change;
- Global Warming
- Rising Temperatures

- Tropical Cyclones / Hurricanes

Each topic of climate change will use a different style of data visualisation, showing the diversity of python's Matplotlib and Seaborn libaries.

### Flooding and Precipitation: Lines and Points

```python
sns.set(rc={'figure.figsize':(6,4)}) 
sns.set_color_codes("dark")
sns.set_style("white")
```

```python
floods = pd.read_csv(r"seaborn_data\flooding_events\Global Flood Archive.csv", encoding = "ISO-8859-1")
```

```python
floods.head()
```

```python
floods = (floods
          .assign(year=lambda floods: floods.Began.str.split('/', expand=True)[2])
          .assign(yearly_cause= lambda floods: floods.groupby(['year', 'MainCause'])['MainCause'].transform('count'),
                  total=lambda floods: floods.groupby(['year'])['ID'].transform('count'),
                  MainCause=lambda floods: floods.MainCause.str.title().str.replace('Brief ',''))
          .assign(cause_percentage= lambda floods: floods.yearly_cause/floods.total*100)
          .query('MainCause == "Heavy Rain" | MainCause == "Monsoonal Rain" | MainCause == "Torrential Rain"')
          .assign(length=lambda floods: floods.Endedn - floods.Begann)
          .astype({'year': 'int'})
          .set_index('year')
          .assign(decade=lambda floods: (floods.index // 10 * 10).astype('str')+'s',
                  year = lambda floods: floods.index))
```

```python
data = (floods[['decade', 'MainCause', 'cause_percentage']]
        .groupby(['MainCause', 'decade'])
        .mean()
        .reset_index()
        .pivot(index='decade', columns='MainCause', values='cause_percentage'))
```

```python
data.head()
```

```python
# create the figure and axes
fig, ax = plt.subplots()

# plot lines
kind = ['Heavy Rain', 'Monsoonal Rain', 'Torrential Rain']
colors = ['skyblue', 'steelblue', 'b']

for k, c in zip(kind, colors):
    sns.pointplot(data=data, x=data.index, y=k, color=c, label=k, ax=ax)

# set title and axis labels
ax.set_title('Causes of flooding events worldwide from 1985 to 2020, by decade', fontdict=title_font, x=-0.125, y=1.05,ha='left')
ax.set_ylabel('Percentage of total flooding events (%)', fontdict=label_font)
ax.set_xlabel('Decade', fontdict=label_font)

# adjust axis, grid lines
sns.despine(left=True)
ax.spines['bottom'].set_linewidth(.5)
sns.despine(left=True)

# adjust parameters
ax.tick_params(colors='dimgrey', labelsize=8.5)

# plot legend
legend_font = fm.FontProperties(family='Trebuchet MS', size=8)
ax.legend(bbox_to_anchor=(0.5, -0.2), loc='center', ncol=6, frameon=False, prop=legend_font)

# add '%' symbol to y-axis tick labels
tick_labels = [f'{int(label)}%' for label in ax.get_yticks().round(2)]
ax.set_yticklabels(tick_labels)

# heavy rain annotations
ax.annotate(str((ax.lines[0].get_xydata()[0][1]).round(2)),
            xy=ax.lines[0].get_xydata()[0],
            xytext=(-9, -15),
            textcoords='offset points',
            fontsize=7, color='skyblue', fontweight='bold')

ax.annotate(str((ax.lines[0].get_xydata()[-1][1]).round(2)),
             xy=ax.lines[0].get_xydata()[-1],
             xytext=(-8, -13),
             textcoords='offset points',
             fontsize=7, color='skyblue', fontweight='bold')

# mosoonal rain annotations
ax.annotate(str((ax.lines[6].get_xydata()[0][1]).round(2)),
            xy=ax.lines[6].get_xydata()[0],
            xytext=(-4.5, 10),
            textcoords='offset points',
            fontsize=7, color='steelblue', fontweight='bold')

ax.annotate(str((ax.lines[6].get_xydata()[-1][1]).round(2)),
            xy=ax.lines[6].get_xydata()[-1],
            xytext=(-6, -15),
            textcoords='offset points',
            fontsize=7, color='steelblue', fontweight='bold')

# torrential rain annotations
ax.annotate(str((ax.lines[12].get_xydata()[0][1]).round(2)),
            xy=ax.lines[12].get_xydata()[0],
            xytext=(-4.5, 10),
            textcoords='offset points',
            fontsize=7, color='b', fontweight='bold')

ax.annotate(str((ax.lines[12].get_xydata()[-1][1]).round(2)),
            xy=ax.lines[12].get_xydata()[-1],
            xytext=(-6, -15),
            textcoords='offset points',
            fontsize=7, color='b', fontweight='bold')

plt.savefig('seaborn_figures/Causes of flooding events worldwide from 1985 to 2020, by decade.jpeg',bbox_inches='tight')
plt.show()
```

```python
data = (floods
        .query('year<=2020')
        .astype({'year': 'int', 'Severity': 'float'})
        .reset_index(drop=True)
        .groupby('year', as_index=False)['Severity'].mean())
```

```python
data.head()
```

```python
# create the figure and axes
fig, ax = plt.subplots()

# plot reg plot
sns.regplot(data=data, x='year', y='Severity',  color='b', ax=ax)

# set title and axis labels
ax.set_title('Severity of flooding events worldwide from 1985 to 2022', fontdict=title_font, x=-0.125, y=1.05,ha='left')
ax.set_ylabel('Severity of flooding event\nbased on yearly interval severity index', fontdict=label_font)
ax.set_xlabel('Decade', fontdict=label_font)

# adjust axis, grid lines
ax.grid(True, axis='y', linestyle='--', linewidth=.1, color='grey')
sns.despine(left=True)
ax.spines['bottom'].set_linewidth(.5)

# adjust parameters
ax.set_xticklabels(ax.get_xticklabels())
ax.tick_params(colors='dimgrey', labelsize=8.5)

# add 's' symbol to x-axis tick labels
tick_labels = [f'{int(label)}s' for label in ax.get_xticks()]
ax.set_xticklabels(tick_labels)

plt.savefig('seaborn_figures/Severity of flooding events worldwide from 1985 to 2022.jpeg',bbox_inches='tight')
plt.show()
```
